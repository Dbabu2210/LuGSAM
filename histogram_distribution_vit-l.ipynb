```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv('/content/right lobe scores.csv')
```

```python
df = pd.DataFrame(data)
```

```python
df.columns
```

```python
x=df["right lobe scores"]
```

```python
plt.figure(figsize=(4,4))
plt.hist(x, bins = 10,color='crimson')
plt.xlim(0,1)
plt.ylim(0,100)
plt.xlabel("BPIoU", fontsize=12)
plt.ylabel("counts", fontsize=12)
plt.title("Histogram Distribution of prompt right lobe", fontsize=10)
plt.show
```
